```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open('./assets/miri_paskal.jpeg')
grey_image = img.convert("L")
grey_image_array = np.array(grey_image)
```

```python
plt.imshow(grey_image_array, cmap='gray')
```

```python
# the method takes (x,y) img (numpy matrix) coordinates and returns (x,y,1) homogeneous coordinates (numpy tensor)
def to_homogeneous_coordinates(img):
    return np.ndarray([img, np.ones((img.shape[0], img.shape[1]))])

hom_img = to_homogeneous_coordinates(grey_image_array)
```

```python
def from_homogeneous_coordinates(points):
    return points[:, :-1] / points[:, -1:]
```
