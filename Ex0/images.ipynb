```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

1. Read an image from your local file system into a numpy array.

```python
img = Image.open('./assets/miri_paskal.jpeg')
```

2. Print the dimensions and type of the image (i.e. the numpy array).

```python
img_array = np.array(img)
print("Original RGB Image:")
print("Shape:", img_array.shape)
print("Type:", img_array.dtype)
print(img_array)
```

3. Display the RGB image using matplotlib.

```python
plt.imshow(img_array)
plt.title("Original RGB Image")
plt.show()
```

4. Transform the image into a grayscale image.

```python
grey_image = img.convert("L")
grey_image_array = np.array(grey_image)
```

5. Print the dimensions and type of the grayscale image (i.e. the numpy array).

```python
print("Original RGB Image:")
print("Shape:", grey_image_array.shape)
print("Type:", grey_image_array.dtype)
```

6. Display the grayscale image using matplotlib (Note: make sure the displayed image looks grayscale!).

```python
print(grey_image_array[0,:10])
plt.title("Grey Image")
plt.imshow(grey_image_array, cmap="gray")
```

7. Save the grayscale image to the disk (Note: use the image processing library and not matplotlib,
matplotlib will reduce the image quality, use it for plotting but avoid using it for saving actual
images).

```python
grey_image.save("./assets/grey_miri_paskal.jpg")
```

8. Using the original, RGB image, transform the range of values from [0, 255] with type uint8 to [0,
1] with type float32.

```python
rgb_array_float32 = img_array.astype(np.float32) / 255.0  # Transform range to [0, 1] with type float32
rgb_array_float32
```

9. Subtract the mean value of the image from all the pixels of the image.

```python
img_mean = np.mean(rgb_array_float32)
print(img_mean)
sub_img = rgb_array_float32 - np.mean(rgb_array_float32)
print(sub_img)
```

10. Normalize the pixel range back to [0, 1].

```python
def normalize(mat):
    # return (mat - np.min(mat)) / (np.max(mat) - np.min(mat))
    return mat - np.min(mat) 

normalized_sub_img = normalize(sub_img)
print(normalized_sub_img)
print(np.max(normalized_sub_img))
```

11. Transform the pixel range back to [0, 255] with type uint8.

```python
normalized_sub_img_back = (normalized_sub_img * 255).astype(np.uint8)
normalized_sub_img_back 
```

12. In the same figure, display the original RGB image, the grayscale image, and the manipulated RGB
image (Hint: Use matplotlib’s subplots)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img_array)
axes[0].set_title('Original RGB Image')

axes[1].imshow(grey_image_array, cmap='gray')
axes[1].set_title('Grayscale Image')

axes[2].imshow(normalized_sub_img_back)
axes[2].set_title('Manipulated RGB Image')

plt.show()
```
