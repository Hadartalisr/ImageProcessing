8. Using the original, RGB image, transform the range of values from [0, 255] with type uint8 to [0,
1] with type float32.

```python
rgb_array_float32 = img_array.astype(np.float32) / 255.0  # Transform range to [0, 1] with type float32
rgb_array_float32
```

9. Subtract the mean value of the image from all the pixels of the image.

```python
img_mean = np.mean(rgb_array_float32)
print(img_mean)
sub_img = rgb_array_float32 - np.mean(rgb_array_float32)
print(sub_img)
```

10. Normalize the pixel range back to [0, 1].

```python
def normalize(mat):
    # return (mat - np.min(mat)) / (np.max(mat) - np.min(mat))
    return mat - np.min(mat) 

normalized_sub_img = normalize(sub_img)
print(normalized_sub_img)
print(np.max(normalized_sub_img))
```

11. Transform the pixel range back to [0, 255] with type uint8.

```python
normalized_sub_img_back = (normalized_sub_img * 255).astype(np.uint8)
normalized_sub_img_back 
```

```python
def compute_fft_2d(data):
    """
    Compute the 2D FFT of a NumPy array.

    Parameters:
    - data: 2D NumPy array

    Returns:
    - fft_result: 2D FFT of the input array
    """
    fft_result = np.fft.fft2(data)
    return fft_result


def plot_fft_magnitude_spectrum(image_fft):
    """
    Display the magnitude spectrum of a 2D FFT as an image.

    Parameters:
    - image_fft: Complex-valued 2D NumPy array (result of FFT)

    Returns:
    - None (displays the plot)
    """
    # Compute the magnitude spectrum
    magnitude_spectrum = np.abs(image_fft)

    # Shift the zero frequency components to the center
    shifted_spectrum = np.fft.fftshift(magnitude_spectrum)

    # Apply log scale for better visualization (avoid very small values dominating)
    log_spectrum = np.log(1 + shifted_spectrum)

    # Display the result as an image
    plt.imshow(log_spectrum, cmap='gray')
    plt.title('Log Magnitude Spectrum')
    plt.colorbar(label='Log Magnitude')
    plt.show()

FFT = compute_fft_2d(grey_image_array)
plot_fft_magnitude_spectrum(FFT)
```
